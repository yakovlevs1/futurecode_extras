# **Модуль**

Программы могут содержать огромное количество строк кода - сотни, тысячи, даже миллионы. В таких случаях невозможно уместить весь код в одном файле, поэтому его необходимо разбивать на отдельные части. Для доступа к коду, находящемуся в других файлах, используется механизм модулей в Python.

**Модуль** в Python - это обычный файл, с расширением `.py`, содержащий код на Python. Каждый модуль может содержать переменные, классы и функции. Кроме того, в модуле может находиться исполняемый код. Имя модуля соответствует имени файла. Например: `math.py`  - это модуль с именем `math`

Каждая программа может импортировать модуль и получить доступ к его классам, функциям и объектам. Модуль может быть написан не только на Python(обычные модули), а например, на C или C++(модули расширения).

# **Пакет**

Если проект состоит из нескольких модулей, то можно объединить их в группы с помощью пакетов (packages) в Python. Пакеты позволяют группировать близкие по тематике модули или выносить часть модулей из проекта для использования в других проектах.

19.png

**Пакет** - это набор взаимосвязанных модулей(сам пакет тоже является модулем), предназначенных для решения задач определенного класса некоторой предметной области. Это такой способ структуризации модулей. Пакет представляет собой папку, в которой содержатся модули и другие пакеты и обязательный файл `__init__.py`, отвечающий за инициализацию пакета.
Пример содержимого стандартного пакета `json`, который состоит из 4 модулей и файла `__init__.py`:

```
json/
├── __init__.py
├── decoder.py
├── encoder.py
├── scanner.py
└── tool.py
```

# **Инструкция import**

23.png

Для того, чтобы воспользоваться функционалом модуля или пакета необходимо его импортировать внутрь программы. При импорте интерпретатор сначала ищет встроенный модуль с заданным именем. Если такого модуля нет, то идёт поиск файла `<имя_модуля>.py` в списке директорий, определённых в переменной `sys.path`.

Переменная `sys.path` (атрибут path модуля sys) – это список строк, которые указывают пути поиска для модулей. При инициализации данный список включает:
* рабочую директорию скрипта (основного модуля);
* переменную окружения `pythonpath` и пути инсталляции Python.

Если модуль не удается найти, возбуждается исключение `ImportError`.

Посмотреть данный список можно следующим образом:

````python
import sys
sys.path
````

## Синтаксис


`import <модуль/пакет> as <псевдоним>, ...`

После ключевого слова `import` указывается название модуля. Если название модуля слишком длинное, или оно вам не нравится по каким-то другим причинам, то для него можно создать псевдоним с помощью ключевого слова `as`. Одной инструкцией можно подключить несколько модулей, хотя этого не рекомендуется делать, так как это снижает читаемость кода. Правильным тоном считается располагать новый импорт на новой строке.

## Пример 1

````python
# импортируем модули time и random
import time
import random

# получаем текущее время в секундах в формате Unix time
print(f'Текущее время: {time.time()}')

# генерируем рандомное число в диапазоне [0, 1) 
print(f'Случайное число: {random.random()}')

# импортируем модуль math с псевдонимом m и обратимся к числу пи
import math as m
print(f'Число пи {m.pi}')
````

# **Инструкция from ... import**

В случае, если неоходимо импортировать только один или несколько модулей из пакета или одну или несколько функций из модуля можно воспользоваться следующей инструкцией:

`from <модуль/пакет> import <функция/модуль> as <псевдоним>, ...`

Для длинных имен также можно назначить псевдоним, указав его после ключевого слова **as**.

## Пример 2

````python
# импортируем число e и функцию cos() из модуля math
from math import e, cos
print(e)
print(cos(0))
````

Следующий формат инструкции from - так называемый Wildcard импорт позволяет подключить все переменные из модуля:

`from <модуль/пакет> import *`



## Пример 3

````python
# импортируем все функции из модуля math
from math import *
print(sin(1))
````

# **Относительный и абсолютный импорт**

При работе с пакетами может возникнуть ситуация, когда необходимо импортировать подпакет или модуль из другого каталога - соседнего или на уровень выше/ниже. Чтобы интерпретатор Python понял, где искать, ему нужно указать правильный путь в дереве каталогов к модулю/пакету, который вы собираетесь импортировать. В Python данный путь может быть двух видов:
* абсолютный
* относительный

При использовании **абсолютного** импорта необходимо указывать полный путь до модуля, начиная с корневого пакета проекта.

## Пример 4
Для проекта `courses_project` укажем адреса модулей `python.py` и `students.py`

```
courses_project
    ├── courses
          ├── __init__.py
          ├── python.py
          ├── c++.py
          └── kotlin.py
    ├── students.py
    └── professional.py
```
*Решение:*

courses.python.py

students.py

При абсолютном импорте одинаково могут использоваться обе выше рассмотренные инструкции:
* **import** <полный_путь>
* **from** <полный_путь> **import** <объект_импорта>


При **относительном** виде импорта используются относительные пути к импортируемому файлу. Относительный путь показывает, на сколько директорий вверх необходимо подняться. Формат относительного пути: 

`.<модуль/пакет>`

Количество точек в начале указывает, как необходимо перемещаться по директориям пакета.
* одна точка . означает, что пакет/модуль лежит в текущей директории. 
*две точки .. говорят о том, что необходимо подняться на уровень выше и так далее.

При относительном импорте(в отличие от абсолютного) может использоваться только вторая инструкция:

`from <относительный путь> import <объект импорта>`

````python
# импортируем модуль python в модуль kotlin(находятся в одном пакете courses)
from . import python

# импортируем функцию hello из модуля python в модуль kotlin(находятся в одном пакете courses)
from .python import hello
````

По стандарту PEP 8 абсолютные импорты предпочтительнее относительных.

**Важно!** Любой скрипт с относительными импортами НЕЛЬЗЯ запустить напрямую. Если это сделать, то неминуемо появится ошибка  `'ImportError: attempted relative import with no known parent package.'`.

# Функции, неформальное пояснение
В чем идея их использования? 

Разобьем программу на некоторый набор блоков, каждый из которых выполняет определённую роль. Зачем?

 - Самое очевидное, частно какой-то код необходимо использовать многократно. Вместо того, чтобы для каждой ситуации копипастить одинковый код, гораздо проще представить его в виде "черного ящика", который получает на вход какие-то данные и делает то, что нам нужно. 

 - Проще и следить за тем, что выполняется в программе, и отлаживать её (потому что мы можем проверять работоспособность и правильность каждого блока по-отдельности, вместо того, чтобы каждый раз искать ошибки во всей программе).

 - Кроме того, никто в реальности не пишет какой-то серьезный проект одним "файлом" с кодом, как это делаем мы во время обучения. Большой проект задействует много различных сущностей, в том числе файлов программ, и самый актуальный способ использовать какой-то код из другого файла - это импортировать функцию или класс (про них позже) оттуда. Так же, как мы делали с библиотеками.

Теперь давайте посмотрим на примерах, как писать и использовать (вызывать) функции.

В любом нормальном проекте должна быть документация, и любой человек, читающий ваш код, должен понимать, что делает та или иная функция.
Для этого по-хорошему нужно писать комментарии и описания к функциям. Конечно, в рамках маленьких задач обучающего курса это бесполезное занятие, но как только вы пишете хотя бы сотню строк, без комментариев становится немного грустно.

Что нам обычно важно знать про функцию?
1) Какие аргументы функция принимает
2) Что она вообще делает и возвращает

```python
"""
Функция принимает:
    число x типа int, 
    еще какой-то аргумент, который не используется никак (не зря он "тупой" в названии), 
    и строку s.

Функция возвращает строку, которая получается умножением строки s на число x. 

На случай отрицательных x использована стандартная функция abs(x)
"""


def my_best_function(x, some_stupid_arg, s):
    answer = abs(x) * s
    # return abs(x) * s
    return answer


# Вызовем функцию и выведем то, что она вернёт
# Замечу, что аргумент посередине обязателен, хотя и бесполезен, об этом факте ниже, см. вызов функции.
print(my_best_function(-10, 1, "&"))
```

Давайте подробнее о том, что произошло в коде выше.

Чтобы обозначить, что мы объявляем функцию, пишется ключевое слово `def`.

Через пробел от него пишется название функции. Для названия функции действуют те же правила, что и для переменных (не должно начинаться с цифры, содержит большие/маленькие буквы и нижнее подчеркивание итд). PEP-8 говорит нам, что, если название функции содержит несколько слов, их надо написать snake-case'ом, то есть маленькими буквами через нижнее подчеркивание.

В круглых скобках после названия через запятую мы объявляем аргументы нашей функции.

Далее идёт стандартная конструкция - двоеточие и перенос строки, а тело функции (то есть всё, что эта функция делает) пишется с одиночным отступом (tab или 4 пробела).

Функция может возвращать какие-либо значения, а может и не возвращать. Для того, чтобы вернуть что-то из функции, нужно написать оператор `return` и через пробел от него то, что необходимо вернуть. Обратите внимание, что возвращение чего-либо завершает выполнение функции, и никакой код внутри функции после этого выполняться не будет. Если `return`'а у функции нет, или после этого оператора ничего нет, то возвращается значение `None`, то есть *ничего*.

Вызывать функции мы уже умеем: для этого достаточно написать название функции и в круглых скобках через запятую отправить в неё аргументы. **Важно**, что по стандарту ВСЕ аргументы являются обязательными, то есть функция от 3 аргументов не будет работать, если отправить только 2 (думаю, не стоит говорить, что если отправить больше, то это тоже не сработает). Кроме того, аргументы являются *позиционными*, то есть всегда передаются в строгом порядке. Ситуация поменяется, как только мы научимся использовать значения по умолчанию, но пока мыслим таким образом. 

Ниже пример ошибки, которую выдаст интерпретатор, если отправить в функцию недостаточное количество аргументов.

```python
print(my_best_function(1, 2))
```

## Области видимости

### Локальная область видимости

Важная особенность функций как элемента языка - область видимости переменных. Всё, чему мы без специальных инструкций (то есть ключевых слов `global` и `nonlocal`, о них ниже) присваиваем какие-то значения, существует только внутри функции. После выхода из функции эти данные утрачиваются и посмотреть на них "извне" не получится.

**Пример.** 

Создадим функцию, которая для каждого элемента списка печатает его квадрат в stdout.

```python
"""
Функция принимает список из чисел и печатает квадрат каждого числа в отдельной строке.
"""


def print_squared_list(lst):
    for element in lst:
        print(element**2)
```

В данной функции переменные `element` и `lst` являются локальными и если мы попробуем к ним обратиться в основном теле программы - нарвёмся на `NameError`

```python
print(element)
```

А если вдруг у нас до выполнения функции были переменные с такими же названиями, от выполнения функции ничего не поменяется.

**Пример**, подтверждающий этот факт:

```python
mas = [9348, 34985, 576, 574892]
element = 11111111111111111111
print_squared_list(mas)  # эту функцию мы объявили ранее
print(element)
```

Итак, переменные внутри функции являются *локальными* и не могут использоваться за пределами своей функции. Если вдруг для какой-либо переменной в другой области видимости существует переменная с тем же названием, то их значения являются независимыми и никак не пересекаются (потому что несмотря на одинаковые названия, ссылаются они на разные объекты в памяти).

Напомню, локальные переменные функции существуют только во время её выполнения и уничтожаются после этого.

### Глобальная область видимости

Помимо локальной области видимости существует глобальная. Когда мы создаем переменные в основной части программы (вне функций), они являются глобальными. К таким переменным можно получить доступ из любой части программы. Но не всё так легко, стоит определиться, что мы можем делать с глобальными переменными, а чего мы **не** можем.

#### Начнём с хорошего. 

Что можно делать? "Смотреть" на какие-то значения и/или сравнивать их с чем-то.

Рассмотрим это на примере.

```python
"""
Функция проверяет список lst на соответствие каким-то значениям из списка reference_data, определенным вне этой функции (в глобальной области видимости)
Возвращает True, если все значения совпадают и False иначе.
"""


def data_check(lst):
    for i in range(len(lst)):
        if lst[i] != reference_data[i]:
            return False
    return True


reference_data = ["Python3.7.3", 101, [1, 9]]

print(data_check(["C++", 101, [1, 9]]))
```

Как вы видите, в этом примере список `reference_data` объявлен в основной части кода, что делает его глобальным для доступа к данным, и наша функция сравнивает значения локального списка `lst` со значениями, которые внутри неё не объявлены.

#### Перейдем к "плохому". 

Значения нельзя менять. Точнее говоря, что-то изменять мы можем, например, в официальной методичке есть пример с функцией, которая удаляет элемент из глобального списка с помощью `del`, но мы благополучно проигнорируем этот факт в угоду красоте изложения (а также нормальным паттернам написания кода) и скажем, что глобальные значения из функции мы менять не можем.

Зададим два резонных вопроса:
1) А если всё же попробовать это сделать?
2) А если очень хочется?

С ними разберемся по порядку. 

##### Что будет, если попробовать изменить глобальные значения изнутри функции?

Рассмотрим **пример**.

```python
"""
Функция просто переприсваивает значения списка на новые.
"""


def new_list():
    some_list = [1, 2, 3]


some_list = [4, 5, 6]
new_list()
print(some_list)
```

Как видно, попытка изменить глобальные значения не привела к ошибке от интерпретатора, но при этом не увенчалась успехом. Если мы используем оператор присваивания (или что-то эквивалентное ему) внутри функции на переменную, которая является глобальной, то внутри списка создается локальная копия этой переменной, которая, как мы знаем, умирает в конце выполнения функции.

##### Теперь второй вопрос - что делать, если нам оооочень нужно поменять какие-то глобальные значения изнутри функции?

В таких ситуациях есть выход - ключевое слово `global`. Во время выполнения функции мы должны явно обозначить, какие переменные будут считаться глобальными (а значит, их значения будут меняться функцией).

**Пример.**

```python
"""
В функции используется ключевое слово global, чтобы обозначить, что глобальное значение x будет изменяться.
Функция удваивает число x.
"""


def double():
    global x
    x *= 2


x = 7
double()
print(x)
double()
print(x)
double()
print(x)
double()
print(x)
```

Хочу отметить, что так лучше не делать. Код с использованием ключевого слова global становится гораздо менее читабельным, так что пользоваться возможностью менять глобальные значения стоит только в крайних случаях: либо когда без этого не обойтись, либо когда код без этого будет сильно сложнее.

## Значения по умолчанию. Именованные аргументы - непозиционные аргументы.

В питоне есть возможность установить значение по умолчанию для аргумента функции. В случае, если аргумент, имеющий значение по умолчанию, не будет передан в функцию, все равно создастся локальная переменная с нужным названием и значением, указанным по умолчанию.

**Рассмотрим на примере.**

Посчитаем чаевые со счета в ресторане. Нормальные люди в среднем оставляют 10% от счета, это значение и сделаем по умолчанию. Обслуживание нам может как очень понравиться, так и очень не понравиться, поэтому иметь возможность менять эти проценты хочется, так что мотивация использовать проценты как аргумент очевидна.

```python
"""
Функция принимает на вход обязательным аргументом значение счета, а необязательным - процент чаевых, которые мы хотим заплатить
Печатает в stdout предложение, содержащее переданные ей параметры и размер чаевых, которые мы заплатим.
"""


def tip(bill_amount, percent=10):
    print(
        f"Для счета в {bill_amount}р чаевые в размере {percent} процентов равны {int(bill_amount * percent / 100)}р.",
        end="\n\n",
    )


# Сначала вызовем функцию по классике
print("tip(2500, 15)")
tip(2500, 15)

# А теперь попробуем новый способ - если у аргументов объявлены значения по умолчанию и мы не хотим их менять,
#  на их место можно ничего не отправлять
print("tip(4500)")
tip(4500)

# Странно это делать для функции с одним значением по умолчанию, но к аргументам, имеющих значение по умолчанию,
#  мы можем обратиться "по имени"
print("tip(11000, percent=11)")
tip(11000, percent=11)
```

**Замечания к примеру**:
 - По стандарту `PEP-8` знаки равенства не отделяются пробелами ни когда мы объявляем стандартное значение в описании функции, ни когда мы обращаемся к именованному аргументу
 - Учитывая, что мы можем иметь несколько аргументов со значениями по умолчанию, какие-то из них можно не использовать при вызове (очевидно, ведь мы ради этого и устанавливали значения по умолчанию), а к каким-то обращаться по имени. Это сильно упрощает нам жизнь: если ранее все аргументы у нас были позиционными, то теперь у нас есть возможность изменить ситуацию. Этот факт поясним примером ниже.
 - **Нет никакого смысла передавать в функцию аргумент, равный своему значению по умолчанию, поэтому никто так не делает. И вы не делайте.**

**Пример** для пояснения второго замечания.

Усложним нашу функцию чаевых, добавим учет скидки ресторана (мало ли мы являемся постоянным клиентами) и возможность оплатить налог за нашего официанта.

```python
"""
Теперь наша функция принимает не только размер нашего счета и процент чаевых, но и скидку для нас как для клиента, 
    а также boolean значение - хотим ли мы заплатить налог за официанта.
Всё так же печатаем предложение со всеми параметрами и результатом.
"""


def tip_advanced(bill_amount, percent=10, discount=0, pay_tax=False):
    final_bill = int(bill_amount * (1 - discount / 100))
    if pay_tax:
        final_tip = int(final_bill * percent / 100 / 0.87)
    else:
        final_tip = int(final_bill * percent / 100)
    print(
        f"Для счета в {bill_amount}р мы получили скидку {discount}% и заплатили {final_bill}р."
    )
    print(
        f"А также оставили чаевые в размере {percent}%, равные {final_tip}р.",
        end="\n\n",
    )


print("tip_advanced(1500, discount=5)")
tip_advanced(1500, discount=5)

print("tip_advanced(850, pay_tax=True)")
tip_advanced(850, pay_tax=True)

# Специально поменял изначальный порядок аргументов - можете заметить, в объявлении функции сначала стоял discount.
print("tip_advanced(1000, pay_tax=True, discount=1)")
tip_advanced(1000, pay_tax=True, discount=1)

print("tip_advanced(1111, 15, 1)")
tip_advanced(1111, 15, 1)
```

Как видно, к именованным аргументам мы можем обращаться в любом порядке, если отдельно обозначаем, какому именно аргументу присваиваем значение. При этом остается возможность просто перечислить значения без указания имен, но стоит помнить, что в таком случае порядок, в котором они передаются в локальные переменные, будет таким же, как и в объявлении функции.

**Замечание**: именованные (или имеющие значения по умолчанию) аргументы должны стоять ПОСЛЕ всех позиционных. Думаю, вы это и сами поймете после того, как пару раз напишете иначе и почитаете ошибку.

## Функции с переменным числом аргументов. Распаковка.

Поскольку это не входит в программу курса, не будем останавливаться на распаковке подробно. Для нас скорее важна возможность использовать ее как инструмент.

Для нас важен лишь тот факт, что неограниченное количество значений, переданных через запятую, "упаковываются" в одну переменную. Эта переменная всегда будет являться кортежем (tuple) - неизменяемым списком.

**Пример 1.** 

Для заранее неизвестного числа сотрудников посчитаем квартальный бонус к зарплате. Стандартным значением бонуса выберем 20%. Вернём список из соответствующих премий.

```python
"""
Неограниченное количество значений ЧЕРЕЗ ЗАПЯТУЮ, то есть различных аргументов, превращаются в один кортеж n, 
по которому мы можем итерироваться, чтобы получить какой-то результат.
"""


def bonus(*n, percent=20):
    return [element * percent // 100 for element in n]


print(bonus(3000, 2000, 2300, 8100, 2500, percent=50))
```

Часто возникают ситуации, в которых функция написана для произвольного числа аргументов, а данные у нас хранятся в списке (ну или другом `iterable` объекте).

Очевидно, что в таком случае нельзя просто отправить список как аргумент. Для того, чтобы всё работало, используем еще одну распаковку.

**Пример 2.**

```python
list_of_salaries = [3000, 2000, 2300, 8100, 2500]

# функция определена ячейкой выше, не будем её никак менять.
print(bonus(*list_of_salaries, percent=50))
```

Таким образом, с помощью распаковки мы обобщили нашу функцию: теперь в неё можно отправить и список (не забываем про звездочку), и просто любое количество аргументов через запятую.

## Функции высшего порядка.

С функциями высшего порядка все просто - это функции, которые принимают в качестве аргумента другие функции. Заметьте, именно функции (точнее ссылки на них), а не возвращаемые ими значения - то есть нужно писать не вызов функции (со скобочками), а просто название нужной нам функции.

В рамках текущего курса нам в основном нужно познакомиться с несколькими важными примерами таких функций.

### map()

Функция высшего порядка `map(function, iterable)` применяет функцию, которую мы в нее отправляем, ко всем элементам итерабельного объекта (то есть того, по которому мы можем пройтись поэлементно) и возвращает объект типа `map` (на самом деле особо не важно, что это, главное, что мы можем это конвертировать в удобный для нас `list`)

**Пример.**

Поскольку мы должны отправить в `map` какую-либо функцию, давайте попутно ее и напишем (нам ничего не мешает использовать какую-то стандартную, но так же не интересно, правда?)

```python
def boom(x):
    return x**5 + 1


lst = [1, 9, 2, 8, 3, 7, 4, 6, 5, 100, 999]
result = map(boom, lst)
print(type(result))
print(result)
print(list(result))
```

**Пояснение** к примеру:

Без какой-либо идеи создадим функцию `boom`, которая принимает число и возвращает его же, возведенное в 5 степень и с прибавленной единичкой. Создадим произвольный список `lst` и применим с помощью `map` к нему функцию `boom`.

Обратите внимание на то, как пишется вызов `map` как функции высшего порядка. 

Для демонстрации код выводит тип данных, который возвращает `map` и само значение (можете заметить, что у класса `map` нет встроенного поведения при выводе, но об этом поговорим, когда речь пойдет о классах). Чтобы всё же красиво продемонстрировать результат конвертируем результат в `list` и распечатаем.

### filter()

Совершенно неожиданно для нас, функция `filter(function, iterable)` *фильтрует* то, что мы в нее отправим, в согласии с функцией, которую мы в нее отправим.

Поясним немного этот момент до примера: функция `function` должна уметь принимать элемент из `iterable` и возвращать значение типа `bool` - эта функция и будет критерием фильтрации. Если для элемента было возвращено `True`, он "проходит проверку" и остается, а иначе - выбывает.

**Пример.**

```python
def some_criteria(k):
    return 99 < k < 1000


lst_2 = [100, 99, 23, 854, 3, 747, 42455, 6123, 532, 100, 999]
result = filter(some_criteria, lst_2)
print(type(result))
print(result)
print(list(result))
```

Пояснение к примеру:

Cоздадим функцию `some_criteria`, которая принимает число и проверяет, что оно трехзначное. Создадим произвольный список `lst_2` и отфильтруем из него трёхзначные числа с помощью `filter`.

Для демонстрации код выводит тип данных, который возвращает `filter` и само значение (можете заметить, что у класса `filter` нет встроенного поведения при выводе, но об этом поговорим, когда речь пойдет о классах). Чтобы всё же красиво продемонстрировать результат конвертируем результат в `list` и распечатаем.

# Type hints или аннотации типов

Часто бывает удобнее напрямую указывать типы данных, которые мы используем в переменных или при вызовах функций. 

Да, мы прекрасно знаем, что Python - язык с сильной динамической типизацией, но так тоже можно. Зачем?

В качестве примера могу привести самую первую функцию, написанную в этой методичке - `my_best_function`. Без комментария разобраться, что за переменные туда передаются, достаточно сложно, и уж точно невозможно с первого взгляда. Казалось бы, это ошибка человека, написавшего функцию, потому что названия переменных могли бы быть и понятнее при прочтении кода. Но в современности мы хотим пользоваться кучей вспомогательной автоматики (умным отслеживанием кода в IDE, линтерами, тайп-чекерами и прочим), а для любой автоматики названия переменных все едины и одинаково непонятны. Как раз для того, чтобы автоматика лучше отслеживала то, как мы пишем код, и внедрена возможность указывать тип данных при объявлении функции.

Нужно отметить, что тайп-аннотации — это именно возможность, а не обязанность. У программиста есть выбор — добавлять информацию о типах или нет. Таким образом Python пытается усидеть на двух стульях — остаться языком с динамической типизацией и дать возможность для статического анализа написанных программ.

### Как это работает?

Программист при написании кода расставляет информацию о типах переменных, параметров и возвращаемых значений функций. Это никак не влияет на выполнение программы. Python сам по себе никак не использует эту информацию в во время выполнения программы, он лишь перекладывает её в специальные атрибуты функций или переменных, делая доступной для сторонних утилит. То есть, если указано, что функция принимает строки, то это никак не помешает вызвать её с целыми числами или списками — в зависимости от тела функции, она может отработать, а может завершиться ошибкой, но сама возможность вызова с любыми типами аргументов никак не ограничивается.

**Зачем же тогда писать тайп-аннотации?**

 - Документация для разработчиков (которую, кстати, очень удобно показывают современные IDE)
 - Сторонние утилиты и библиотеки могут использовать эту информацию по своему усмотрению, например, выполняя проверки типов или приводя данные к нужному типу.

### Примеры.

**Пример 1 - простая аннотация типа данных переменной**

```python
name: str = "Alex"
age: int = 21
is_married: bool = False  # пока что


x: float  # Можно аннотировать переменную, не присваивая какого-либо значения.
```

**Пример 2 - несколько вариантов аннотации типов у функции**

```python
# Просто указываем предполагаемый тип данных аргумента
def greeting(name: str):
    return "Hello, " + name


# То же самое, но со стандартным значением для этого аргумента
def greeting(name: str = "world"):
    return "Hello, " + name


# То же самое, но добавим предполагаемый тип данных возвращаемого значения
def greeting(name: str = "world") -> str:
    return "Hello, " + name


# Аналогично, просто другой пример
def body_mass_index(weight: float, height: float) -> float:
    return weight / height**2
```

# Лямбда-функции. Lambda. Анонимные функции.

Теперь мы знаем про функции высшего порядка. В них используются функции первого порядка, которые имеют достаточно простую структуру: взять элемент - сделать небольшое действие - вернуть какое-либо значение. И тем не менее, это должны быть именно функции. Часто не хочется ради этого идти и отдельно объявлять функцию, чтобы потом ее использовать. Проще объявить так называемую *анонимную* функцию прямо "на месте", там, где она применяется.

Посмотрим на пару **примеров**.

```python
# Пояснения к коду ниже
def criteria(x):
    return x > 373


m = [1, 3, 56546, 345, 23465, 654, 374, 546456, 123, 345]
res_1 = list(filter(criteria, m))
res_2 = list(filter(lambda w: w > 373, m))
print(res_1 == res_2)
print(res_2)
```

**Поясним**, что произошло и как это написано. 

Для демонстрации была выбрана функция `filter`, но происходящее легко обобщается на любую ситуацию. 

Как и ранее, создадим функцию с любым названием (в моем случае `criteria()`), создадим список для фильтрации и вызовем функцию `filter`. Результат `res_1` конвертируем в `list` для удобства.

Далее вместо того, чтобы использовать заранее объявленную функцию, объявим её прямо в момент вызова `filter`. Результат запишем в другую переменную `res_2`, но, как видно, результаты работы в обоих случаях одинаковы.

Теперь поговорим про **синтаксис**. Чтобы объявить лямбда-функцию, мы должны написать ключевое слово `lambda`, указать, что она принимает, поставить двоеточие и после него то, что она возвращает.

`lambda <аргумент(-ы)>: <что вернуть>`

Аргументов может быть много, а возвращаемое значение может иметь самую разную стуктуру (собственно, как и у обычных функций). Важно то, что лямбда всегда что-то возвращает.

Также лямбда-функцию можно куда-нибудь присвоить, это нам пригодится в контестах, хотя достаточно странно использовать что-то такое в реальной практике, ибо от обычной функции это становится неотличимо (хотя при этом на нас накладываются все ограничения лямбда-функций).

```python
some_function = lambda x: [int(i) for i in str(x)]  # можно даже так

print(some_function(2984357564564545645456))
```

**Еще один пример.**

Отсортируем список строк интересным способом:

В обычном сравнении двух строк сравниваются сначала коды первых элементов (а мы помним, что это же эквивалентно лексикографическому порядку из-за того, что кодировки сделаны по-умному), если они равны, то вторые, и так далее.

Мы же изменим этот порядок и в качестве ключа сортировки будем использовать "развёрнутую" строку.

Когда мы напрямую указываем ключ (например, для сортировки `sorted`), это означает, что функция, которую мы указали ключом ***сначала*** применится ко всем элементам, а потом они уже будут сортироваться в зависимости от своего типа данных (в рамках нашего примера только что объяснялось, как сравниваются строки).

```python
str_list = ["Alice", "Bob", "John", "Kate", "Yulia", "Alex", "Nastya"]

result = sorted(str_list, key=lambda s: s[::-1])
print(result)
```

# Рекурсия

Ни для кого не будет удивлением, что внутри функции можно вызывать другие функции. Но в это же правило входит и то, что внутри некоторой функции мы можем вызвать её же. 

Ситуации, в которых это происходит, называются *рекурсией*. 

Для тех, кому хочется подробное описание "на пальцах", рекомендую посмотреть Хирьянова (видео про C, а не Python, но само понятие рекурсии не имеет привязки к языку)

https://www.youtube.com/watch?v=NOaSY5pJmyc

#### Когда использовать рекурсию в задачах и какие вообще задачи можно решить с помощью неё? 

*Задачи, которые можно свести к подзадачам, основанных на исходной задаче, но меньшей размерности.*

Один из простейших **примеров** - факториал. Чтобы узнать $n!$, нужно $(n-1)!$ умножить на $n$. То есть, чтобы узнать факториал $n$, нам нужно узнать $n-1$. Конечная задача подсчета свелась к такой же, но меньшей размерности ("размерность" - обобщение того, насколько много действий необходимо сделать, чтобы решить задачу).

У рекурсии есть один важный аспект - она должна быть конечной. Для этого нужен *крайний случай* (более аккуратное название - *база рекурсии*). Обычно он обрабатывает какую-то элементарную ситуацию. В рамках задачи подсчета факториала это факториал $1! = 1$.
 
Минимальное представление о рекурсивных алгоритмах мы имеем, можем переложить эти знания в практику.

**Пример 1. Факториал.**

```python
def factorial(k: int) -> int:
    if k == 1:
        return 1
    return k * factorial(k - 1)


print(factorial(10))
```

**Поясним**, что происходит в коде. 

 -  Обратите внимание на проверку `k == 1`, это как раз является "крайним случаем", то есть моментом, когда мы уже можем сразу вернуть какой-то результат, не уменьшая размерность задачи. Напомню, наличие крайнего случая необходимо, иначе рекурсия станет (условно) бесконечной.

 -  Если крайний случай не реализуется, сводим задачу к меньшей. В нашем случае для нахождения $k!$ нам нужно найти $(k-1)!$, поэтому вызываем функцию от $k-1$. 
 
 Заметьте, вызов функции не завершится, пока не получит результат выполнения рекурсивного вызова. То есть сначала мы заходим максимально "глубоко" в рекурсивные вызовы (это называется *прямой ход рекурсии*), а только потом возвращаемся и вычисляем дальнейшие значения (это называется *обратный ход рекурсии*).  

 То, насколько "глубоко" мы зашли в рекурсивных вызовах называется ***глубиной рекурсии***. Бывает важно следить за этим параметром когда мы пишем код, потому что возможная глубина рекурсии ограничена программно. В `Python` мы можем посмотреть на это ограничение с помощью `sys.getrecursionlimit()` из модуля `sys`.

**Пример 2. Быстрое возведение в степень.**

Рассмотрим один из классических алгоритмов - быстрое возведение в степень.

Пусть нам необходимо возвести число $t$ в степень $k$.

Чтобы ускорить обычный алгоритм возведения в степень ($k$ раз умножить число $t$ на себя), заметим некоторый факт:

Если нам нужно возвести число в чётную степень $n$, можно упростить (с точки зрения вычислений) наше возведение - вычислить $(t^{n/2})^2$. В сравнении с обычным алгоритмом, каждое подобное действие уменьшает количество вычислений вдвое.

Итого получим следующие рекуррентные соотношения:

$t^k = t^{k-1} \cdot t$, $k$ - нечётное

$t^k = (t^{k/2})^2$, $k$ - чётное

Определимся с базой рекурсии. Очевидно, что возведение в нулевую степень всегда даёт 1, такую ситуацию и выберем в качестве базы.

Обычное возведение в степень работает за $O(n)$, быстрое - за $O(\log n)$

```python
def fast_pwr(t: int, k: int) -> int:
    if k == 0:
        return 1
    if k % 2 == 1:
        return t * fast_pwr(t, k - 1)
    else:
        return fast_pwr(t, k // 2) ** 2


fast_pwr(2, 1000)
```

### Замечания про рекурсию.

 - Рекурсия не добавляет в `Python` (а вообще говоря в любые императивные языки программирования) ничего нового. Нет таких программ, которые могут быть написаны с помощью рекурсии, но не могут быть написать без неё.
 - Рекурсия почти всегда потребляет больше ресурсов (потому что всем вызовам функций необходимо "запоминать" (= записывать в специальную стуктуру - стек вызовов) свое состояние в момент рекурсивного вызова). 
 - Рекурсия без оптимизаций медленнее итераций. Да, это исправляется использованием *хвостовой рекурсии (tail recursion или tail call)* （правда она еще должна поддерживаться компилятором, благо для `Python` это так) и *мемоизации (memoization)*. С оптимизациями разница в лучшем случае упирается в конкретную задачу.
 Хвостовая рекурсия меньше забивает стек. Мемоизация - подход, задействующий *компромисс времени и памяти (space-time trade-off)*, то есть ради ускорения выполнения программы мы используем больше памяти, что снова возвращает нас к тому, что рекурсия прожорлива.

 Зачем тогда использовать рекурсию? 
 
 Многие задачи имеют достаточно простые рекуррентные соотношения, которые легко перенести в код. Код рекурсивных функций в среднем проще аналогичного кода, написанного с помощью цикла. 
 
 Цитируя книгу "Грокаем алгоритмы": "Циклы могут ускорить работу программы. Рекурсия может ускорить работу программиста. Выбирайте, что важнее в вашей ситуации!" 
 
 (кстати, автор(-ка) книги взяла эту цитату из обстуждения на StackOverflow - https://stackoverflow.com/a/72694)

 С моей точки зрения эта цитата полностью отражает суть использования рекурсии: если нужна производительность, предпочтительнее циклы. Если нужно быстро написать код - рекурсия. На двух стульях усидеться не получится: очень сомневаюсь, что написание рекурсии со всеми оптимизациями займет меньше времени, чем написание цикла.


#### TLDR

В императивных языках программирования (к коим относится и `Python`) циклы либо быстрее, либо работают с той же скоростью, но используют меньше памяти. В функциональных языках наоборот.

***Рекурсию быстрее писать, итерации быстрее работают.***
